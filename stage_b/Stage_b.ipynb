```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#load the data into pandas

df = pd.read_csv('energydata_complete.csv')
df.head()
```

```python
#quick description of the data
df.info()
```

```python
df_copy = df.drop(['date', 'lights'], axis = 1) 
df_copy.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df_copy),  columns = df_copy.columns)
X = df_scaled.drop(columns=['Appliances'])
y = df_scaled['Appliances']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print('X_train rows: {} columns: {}'.format(X_train.shape[0], X_train.shape[1]))
print('y_train rows: {}'.format(y_train.shape[0]))
print('X_test rows: {} columns: {}'.format(X_test.shape[0], X_test.shape[1]))
print('y_test rows: {} '.format(y_test.shape[0]))
```

```python
df_reg = df_scaled[['T2', 'T6']]

df_reg.head()
```

```python
x= df_reg['T2'].values.reshape(-1,1)
y = df_reg['T6'].values.reshape(-1,1)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python

from sklearn import linear_model
linear_regression = linear_model.LinearRegression()

linear_regression.fit(xtrain, ytrain)
predicted = linear_regression.predict(xtest)
```

```python
#R-squared or Coefficient of determination
from sklearn.metrics import r2_score

R2_score = r2_score(ytest, predicted)
print('R-squared:',(round(R2_score, 2)))
```

### Linear Model on Main Dataset

```python
from sklearn import linear_model

regression = linear_model.LinearRegression()

regression.fit(X_train, y_train)
energy_predicted = regression.predict(X_test)

print("Training score: {:.3f}".format(regression.score(X_train, y_train)))
print("Test score: {:.3f}".format(regression.score(X_test, y_test)))
```

```python
from sklearn.metrics import mean_absolute_error
MeanAbsoluteError = mean_absolute_error(y_test, energy_predicted)

print('MAE:',(round(MeanAbsoluteError, 2)))
import numpy as np
RSS = np.sum(np.square(y_test - energy_predicted))
print('RSS:',(round(RSS, 2)))

from sklearn.metrics import  mean_squared_error

RMSE = np.sqrt(mean_squared_error(y_test, energy_predicted))
print('RMSE:',(round(RMSE, 3)))

from sklearn.metrics import r2_score

R2_score = r2_score(y_test, energy_predicted)
print('R-squared:',(round(R2_score, 2)))
```

```python
def weights(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
linear_model_weights = weights(regression, X_train, 'Linear_model_Weight')
linear_model_weights
```

```python
linear_model_weights['Linear_model_Weight'].describe()
```

```python
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(X_train, y_train)

ridge_pred = ridge_reg.predict(X_test)

print("Training score: {:.3f}".format(ridge_reg.score(X_train, y_train)))
print("Test score: {:.3f}".format(ridge_reg.score(X_test, y_test)))
```

```python
from sklearn.metrics import  mean_squared_error

RMSE = np.sqrt(mean_squared_error(y_test, ridge_pred))
print('RMSE:',(round(RMSE, 3)))
```

```python
from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(X_train, y_train)

lasso_pred = lasso_reg.predict(X_test)

print("Training score: {:.3f}".format(lasso_reg.score(X_train, y_train)))
print("Test score: {:.3f}".format(lasso_reg.score(X_test, y_test)))
```

```python
lasso_weights_df = weights(lasso_reg, X_train, 'Lasso_weight')
lasso_weights_df[lasso_weights_df["Lasso_weight"] != 0]
```

```python
from sklearn.metrics import  mean_squared_error

RMSE = np.sqrt(mean_squared_error(y_test, lasso_pred))
print('RMSE:',(round(RMSE, 3)))
```
